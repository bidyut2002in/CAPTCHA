```python
def question_generation():
    # csv2: image_no, count, color, class
    # Create cvs3 from csv2: image_no, question, answer
    import csv
    qg=[]
    with open('CSV/shapes_csv2.csv', 'rt')as f:
        data = csv.reader(f)
        for row in data:
            qg.append([row[0],row[1],row[2]])
    f.close()
    file=open('CSV/shapes_csv3.csv', 'w', newline='')
    writer = csv.writer(file)
    for k in qg:
        if int(k[1])>1:
            caption="there are "+ k[1]+" "+str(k[2])+"s"
        elif int(k[1])==1:
            caption="there are "+ k[1]+" "+str(k[2])
        question="how many "+str(k[2])+"s are there?"
        ans=k[1]
        writer.writerow([k[0], caption, question, ans]) 
    file.close() 
    
def object_count():
    # csv1: image_no, x, y, w, h, class
    # create csv2 from csv1: image_no, count, color, class
    import csv
    d={}
    with open('CSV/shapes_csv1.csv', 'rt')as f:
        data = csv.reader(f)
        for row in data:
            if (row[0],row[-1]) not in d:
                d[(row[0],row[-1])] = 0 
            d[(row[0],row[-1])] += 1
    f.close()
    file=open('CSV/shapes_csv2.csv', 'w', newline='')
    writer = csv.writer(file)
    for k in d:
        writer.writerow([k[0],str(d[k]),k[1]]) 
    file.close()             
```
